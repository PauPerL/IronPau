## iterate to asemble all info

```python
imdb_position= []
movie_title= []
release_year= []
imdb_rating= []
dir_star_names= []

len_movies=len(soup.select('td.ratingColumn.imdbRating strong'))
len_movies_check=len(soup.select('td.titleColumn span.secondaryInfo'))
```

```python
print(len_movies , len_movies_check) #check the length of the list and that they match
```

```python
for i in range(len_movies):
    imdb_position.append(i+1) #esto es un poco trampa porque nos estamos inventando la posicion segun hacemos el for
    movie_title.append(soup.select('td.titleColumn a')[i].get_text())
    release_year.append(soup.select('td.titleColumn span.secondaryInfo')[i].get_text())
    imdb_rating.append(soup.select('td.ratingColumn.imdbRating strong')[i].get_text())
    dir_star_names.append(soup.select('td.titleColumn a')[i]['title'])
    
```

```python
#a nice this to see the progres of our code iterations

from tqdm.notebook import tqdm
```

```python
imdb_position= []
movie_title= []
release_year= []
imdb_rating= []
dir_star_names= []

for i in tqdm(range(len_movies)):
    imdb_position.append(i+1) #la idea seria limpiar el input que pillamos de la pagina web
    movie_title.append(soup.select('td.titleColumn a')[i].get_text())
    release_year.append(soup.select('td.titleColumn span.secondaryInfo')[i].get_text())
    imdb_rating.append(soup.select('td.ratingColumn.imdbRating strong')[i].get_text())
    dir_star_names.append(soup.select('td.titleColumn a')[i]['title'])
    
```

## create a clean df of this info

```python
movies_250 = pd.DataFrame({'rank':imdb_position,
                           'title':movie_title,
                           'released':release_year, 
                          'rating':imdb_rating,
                           'people':dir_star_names})
```
