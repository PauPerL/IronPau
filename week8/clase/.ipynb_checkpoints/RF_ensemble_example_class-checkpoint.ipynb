```python
import pandas as pd
import numpy as np
```

# Geting the data

```python
cat=pd.read_csv('categorical.csv')
num=pd.read_csv('numerical.csv')
tar=pd.read_csv('target.csv')
```

```python
data=pd.concat([num,cat,tar],axis=1)
```

```python
data
```

```python
data.columns.to_list()
```

```python
data['TARGET_B'].value_counts()
```

## correcting the imbalance

```python
#downsample using logic -- hecho a lo bruto

category_0= data[data['TARGET_B']==0].sample(len(data[data['TARGET_B']==1]))
```

```python
category_1=(data[data['TARGET_B']==1])
```

```python
#now we gona recreate data?

data=pd.concat([category_0,category_1], axis=0)
data.reset_index(inplace=True,drop=True)
data.shape
```

```python
data['TARGET_B'].value_counts()
```

## pre processing

```python
#split ya
X=data.drop(['TARGET_B'],axis=1)
Y=data['TARGET_B']
```

```python
#hacer categorical y numerical

numericals = X.select_dtypes(np.number)
categoricals= X.select_dtypes(object)
```

```python
### one hot encoder

from sklearn.preprocessing import OneHotEncoder as ohe

encoder = ohe(drop='first').fit(categoricals)   #hace el encoder pero quita una de las columnas que es complementaria al resto
cat_enc=encoder.transform(categoricals).toarray()
cat_enc_cols=encoder.get_feature_names_out(categoricals.columns)
cat_enc_df=pd.DataFrame(cat_enc, columns=cat_enc_cols)
```

```python
#join back the numbers

X_full=pd.concat([cat_enc_df,numericals],axis=1)
```

```python
X_full.shape
```

## train test split and model

```python
from sklearn.model_selection import train_test_split

```

```python
#split train test
X_train,X_test,Y_train,Y_test=train_test_split(X_full,Y,test_size=0.3,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier as rfc
```

```python
# hagamos nuestro modelo de random forest con parametros y lo fiteamos
clf=rfc(n_estimators=50,max_depth=5,random_state=1)
clf.fit(X_train,Y_train)
```

```python
#miremos que tal ha ido

print(clf.score(X_test,Y_test))
```

## cross validation

- proxy test for overfit

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_scores=cross_val_score(clf,X_train,Y_train,cv=10)
np.mean(cross_val_scores)
```

## feature importance

```python
#once we have the rforest we can get the features importance

feature_importance=clf.feature_importances_
```
