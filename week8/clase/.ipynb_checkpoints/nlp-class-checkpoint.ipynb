### 1.3 text cleaning

- remove stop words
- remove punctuation
- lemmatising / stemming
- pronouns, nouns and entities/ abbreviations
- prepositions


```python
#punctuation spaces numbers
clean_text = [word for word in lower_text if word.isalpha()]
```

```python
len(clean_text)
```

```python
#get stop words - appropiate for text
from nltk.corpus import stopwords
```

```python
#lets look at the stop words on english
my_stopwords=stopwords.words('english')
```

```python
my_stopwords=stopwords.words('spanish')
```

```python
print( stopwords.fileids())
```

```python
my_stopwords
```

```python
clean_text = [word for word in clean_text if word not in my_stopwords]
```

```python
len(clean_text)
```

```python
clean_text
```

```python
#count frequency of the words - 2 version
fdist_v2=nltk.FreqDist(clean_text)
word_freq_v2=dict((word,freq) for word, freq in fdist_v2.items())
sorted(word_freq_v2.items(), key=lambda x:x[1], reverse=True)
```

```python
#steaming?

nltk.PorterStemmer.fileids()

```

```python
#this is the steaming but I think its for english lang

porter = nltk.PorterStemmer()
```

```python
stem_words=[porter.stem(word) for word in clean_text]
```

```python
# 3 version of word freq

fdist_v3=nltk.FreqDist(clean_text)
word_freq_v3=dict((word,freq) for word, freq in fdist_v3.items())
sorted(word_freq_v3.items(), key=lambda x:x[1], reverse=True)
```

## 2.1 get sentiment scores

```python
# textblobs -- cant handle nulls


clothdf= clothdf.dropna(subset=['Review Text'], how='any')
```

```python
clothdf.info()
```

```python
clothdf.iloc[2]['Review Text']
```

```python
#get sentiment
example = TextBlob(clothdf.iloc[2]['Review Text'])
example.sentiment
```

```python
def get_sent_p(text):
    return TextBlob(text).sentiment.polarity
```

```python
clothdf['sentiment_p']=clothdf['Review Text'].apply(get_sent_p)
```

```python
def get_subj_p(text):
    return TextBlob(text).sentiment.subjectivity
```

```python
clothdf['subjectivity_p']=clothdf['Review Text'].apply(get_subj_p)
```

```python
clothdf.head()
```

```python
# hagamos otro groupby

clothdf.groupby(['Division Name','Department Name' ]).agg({'Rating':['mean'],'sentiment_p':'mean', 'subjectivity_p':'mean'})
```

```python
#para facilitar analisis hagamos otra columna combinacion
clothdf['dep_div']= clothdf['Division Name']+' - '+clothdf['Department Name']
```

```python
clothdf.groupby('dep_div').agg({'Rating':['mean'],'sentiment_p':['mean','min','max'], 'subjectivity_p':'mean'})
```

```python
person_list=[]

for ent in tokens.ents:
    if ent.label_== 'PERSON':
        person_list.append(ent.text)
```

```python
person_counts=Counter(person_list).most_common(25)
df_person=pd.DataFrame(person_counts,columns=['text','count'])
df_person.plot.barh(x='text', y='count')
```
