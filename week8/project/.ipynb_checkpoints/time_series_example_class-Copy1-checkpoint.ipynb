# rolling avarage

```python
#rolling calcs
rolling_mean=df['temperature'].rolling(window=3).mean()
rolling_std=df['temperature'].rolling(window=3).std()

original=plt.plot(df['temperature'], color='purple', label='original')
mean=plt.plot(rolling_mean, color='orange', label='mean')
#std=plt.plot(rolling_std, color='black', label='std') #no funciona porque es el valor total pero habrai que calcularlo en base al mean

plt.legend(loc='best')
plt.title('rolling man and std')
plt.show();
```

## exponential smoothing

```python
from statsmodels.tsa.api import SimpleExpSmoothing

es= SimpleExpSmoothing(df.temperature.values)
es.fit(smoothing_level=0.1)

plt.plot(df.temperature.values)
plt.plot(es.predict(es.params,start=0,end=None))
plt.show();
```

## ARIMA model

- AR auto regresive model
- I integrated - we dont use the values we use the diference
- MA moving avarege model

```python
# anoying thing about arima
# ARIMA (p d q )-automated ways of deiciding -- pyramid - autoarima

#lets take 1 metric

selected = df['temperature']
```

```python
# de momento escondemos warnings porque si

import warnings
warnings.filterwarnings('ignore')
```

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(selected, order=(2,0,5))
```

```python
#fiteamos
model_fit1=model.fit()
```

```python
df['forecasting_arima']= model_fit1.predict(len(df['temperature'])-20, len(df['temperature'])-1)
```

```python
df
```

```python
df[['forecasting_arima','temperature']].plot(figsize=(15,5));
```

## into the future

```python
df.max()
```

```python
start = datetime.datetime.strptime('2019-01-01', '%Y-%m-%d')
date_list=[start+relativedelta(month=x) for x in range(0,12)]
future=pd.DataFrame(index=date_list,columns=df.columns)
```

```python
future
```

```python
forecast_df=pd.concat([df,future], axis=0)
```

```python
forecast_df
```

```python
forecast_df['future_prediction']=model_fit1.predict(start=35,end=47,dynamic=True)
forecast_df[['temperature','future_prediction']].plot(figsize=(15,5));
plt.grid(True);
```

```python
forecast_df[['temperature','future_prediction']].plot(figsize=(15,5));
plt.grid(True);
```
