```python
df= pd.read_csv('merged_clean_ver1.csv')
```

## clean the gender label! 

```python
df['gender'].unique()
```

```python
# capitalise first letter 
df['gender']=df['gender'].str.title()
```

```python
# case / if logical approach 
def clean_gender(x):
    if x in['M','Male']:
        return 'Male'
    if x in['F','Female','Feamale']:
        return 'Female'
    else:
        return 'Unknown'
```

```python
df['gender']=list(map(clean_gender,df['gender']))
```

```python
df['gender'].unique()
```

```python
corr_data=df.drop(['Unnamed: 0','id','state','gender'],axis=1) # or select_dtypes...
correlation_matrix=corr_data.corr()
correlation_matrix
```

```python
sns.heatmap(correlation_matrix, annot=True);
```

```python
mask=np.zeros_like(correlation_matrix)
mask[np.triu_indices_from(mask)]=True
ax=sns.heatmap(correlation_matrix, mask=mask, annot=True)
plt.show();
```
