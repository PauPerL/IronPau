```python
#we start standarizing the position column
df1['position']=df1['position'].str.lower()
df1['position']
```

```python
df1['position']=df1['position'].str.replace(' ','_', regex=True)
df1['position'].unique()
```

```python
df1.loc[df1['position'].str.contains('data_scientist|machine_learning|statistician'), 'data_scientist']=1
```

```python
df1
```

```python
df1.loc[df1['position'].str.contains('data_analyst|data_analytics|_analytics|quantitative_analyst'), 'data_analyst']=1
```

```python
df1.loc[df1['position'].str.contains('data_engineer|data_science_engineer|quantitative_data_engineer'), 'data_engineer']=1
```

```python
df1
```

```python
df1.duplicated().sum()
```

```python
df1['data_scientist'] = df1['data_scientist'].fillna(0)
df1['data_analyst'] = df1['data_analyst'].fillna(0)
df1['data_engineer'] = df1['data_engineer'].fillna(0)
```

```python
df1.tail()
```

```python
df1.drop('description', axis = 1, inplace = True)
```

```python
df1['numreview']=df1['numreview'].str.replace(' reviews','', regex=True)
```

```python
df1['numreview'] = df1['numreview'].fillna(0)
```

```python
df1
```

```python
df1.info()
```

```python
#we convert columns in to interesting dtypes
df1['data_scientist'] = df1['data_scientist'].astype(int)
df1['data_analyst'] = df1['data_analyst'].astype(int)
df1['data_engineer'] = df1['data_engineer'].astype(int)

#df1["numreview"] = df1["numreview"].astype(float)
```

```python
df1.to_csv('df1_NY.csv')
```
