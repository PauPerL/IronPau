```python
import pandas as pd
```

```python
df = pd.read_csv('web_data.csv')
```

```python
df.dtypes
```

```python
df.head()
```

```python
#convertimos la columna a datetime
df['date_time'] = pd.to_datetime(df['date_time'], errors='coerce') #es bastante capaz de distinguir formato americano de europeo
```

```python
df.info()
```

```python
df.head()
```

```python
#con la columna date_time ya en forma podemos aplicar cosas como esta para sacar una columna que diga el dia
df['day']=pd.DatetimeIndex(df['date_time']).day
```

```python
#mostrando el dato en un formato concreto
df['date_time'][1].strftime(format='%a-%b-%Y')
```

```python
#haciendo un query para sacar entradas que tienen day == 1
df.query('day==1')
```
