```python
import pandas as pd
import numpy as np
```

```python
df_train = pd.read_excel('cookies_comp_students.xlsx', 'cookies_train')
df_test = pd.read_excel('cookies_comp_students.xlsx', 'cookies_test')
```

```python
df_train.info()
```

```python
df_train[df_train['sugar index'].isna()]
```

### dropping nulls

```python
df_train = df_train.dropna()
```

```python
df_train.info()
```

### combining dataframes

```python
df_test = df_test.drop(['id'], axis = 1)
```

```python
df = pd.concat([df_train,df_test], axis = 0)
```

```python
df.reset_index(inplace = True, drop =True)
```

```python
df
```

### split into types

```python
y = df['quality']
X = df.drop(['quality'], axis = 1)
```

```python
numericals = X.select_dtypes(np.number)
categoricals = X.select_dtypes(object)
```

### numericals

```python
numericals.columns
```

```python
# min max scaler
from sklearn.preprocessing import MinMaxScaler
```

```python
num_scaled = MinMaxScaler().fit_transform(numericals.values)
df_numericals = pd.DataFrame(num_scaled, columns = ['sugar to flour ratio', 'sugar index', 'bake temp', 'chill time',
       'calories', 'density', 'pH', 'grams baking soda', 'bake time', 'weight',
       'diameter', 'crunch factor', 'aesthetic appeal'])
```

```python
df_numericals
```

### categoricals

```python
categoricals
```

```python
# one hot encoder
from sklearn.preprocessing import OneHotEncoder as ohe
encoder = ohe(drop = 'first').fit(categoricals)
cat_enc = encoder.transform(categoricals).toarray()
cat_enc_cols = encoder.get_feature_names_out(categoricals.columns)
cat_enc_df = pd.DataFrame(cat_enc, columns = cat_enc_cols)
```

```python
df_categoricals = cat_enc_df
df_categoricals
```

```python
X_full = pd.concat([df_categoricals,df_numericals], axis =1)
```

```python
X_full
```

```python
df_full = pd.concat([X_full,y], axis =1)
```

```python
df_test_final = df_full[df_full['quality'] == 0]
df_test_final
```

```python
df_train_final = df_full[df_full['quality'] != 0]
df_train_final
```

```python
df_train_final['quality'].value_counts()
```

### model

```python
y_train = df_train_final['quality']
X_train = df_train_final.drop(['quality'], axis = 1)
```

```python
from sklearn.ensemble import RandomForestClassifier as rfc
clf = rfc(n_estimators = 101, max_depth = 6, random_state = 40)
clf.fit(X_train, y_train)
```

```python
# feature importance
feature_importances = clf.feature_importances_
feature_names = X_train.columns
forest_importances = pd.Series(feature_importances, index = feature_names)
lst = forest_importances.tolist()
forest_importances.sort_values(ascending = False)
```

### re-run optimized model

```python
y_train = df_train_final['quality']
X_train = df_train_final.drop(['quality'], axis = 1)
```

```python
X_train_opt = X_train[["butter type_melted", "bake temp", "calories", "bake time", "density", "chill time", 
"sugar index","sugar to flour ratio", "grams baking soda", "weight", "pH"]]    
```

```python
X_train_opt
```

```python
y_test = df_test_final['quality']
X_test = df_test_final.drop(['quality'], axis = 1)
```

```python
X_test_opt = X_test[["butter type_melted", "bake temp", "calories", "bake time", "density", "chill time", 
"sugar index","sugar to flour ratio", "grams baking soda", "weight", "pH"]]    
```

```python
X_test_opt.shape
```

```python
# random forrest depth = 6
clf = rfc(n_estimators = 101, max_depth = 6, random_state = 40)
clf.fit(X_train_opt, y_train)
y_test = clf.predict(X_test_opt)
```

```python
# submittion 1 (RF dept = 6)
y_test
```

```python
df_y_test = pd.DataFrame(y_test)
df_y_test.value_counts()
```

```python
# random forrest depth = 8
clf = rfc(n_estimators = 101, max_depth = 8, random_state = 40)
clf.fit(X_train_opt, y_train)
y_test = clf.predict(X_test_opt)
```

```python
# submittion 2 (RF depth = 8)
y_test
```

### depth = 8 is better

```python
# KNN
from sklearn.neighbors import KNeighborsClassifier as knc
clf = knc(n_neighbors = 5)
clf.fit(X_train_opt, y_train)
y_test = clf.predict(X_test_opt)
```

```python
# submittion 3 (k = 5)
y_test
```

```python
# KNN #2
from sklearn.neighbors import KNeighborsClassifier as knc
clf = knc(n_neighbors = 7)
clf.fit(X_train_opt, y_train)
y_test = clf.predict(X_test_opt)
```

```python
y_test
```

### stacking

```python
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression

estimators = [('rf', rfc(n_estimators=101, random_state=40, max_depth = 8)),('knc', knc(n_neighbors = 5))]
clf = StackingClassifier(estimators=estimators, final_estimator= LogisticRegression(max_iter = 200))
clf.fit(X_train_opt, y_train)
y_test = clf.predict(X_test_opt)
```
