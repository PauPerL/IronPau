## now we get our response.content from the website (ie scrape all pages)

```python
pages=[]
for i in tqdm(iterations):
    start_at=str(i)
    url='https://www.imdb.com/search/title/?title_type=feature&release_date=1990-01-01,2021-12-31&user_rating=7.0,10.0&start='+start_at+'&ref_=adv_nxt'
    print(url)
    response =requests.get(url)
    print('status code: '+str(response.status_code))
    pages.append(response.content)
    #have a nap
    wait_time=randint(1,3)
    print('now I will sleep for '+str(wait_time))
    sleep(wait_time)
```

## hagamos lo de las paginas, continuamos con sean

```python
len(pages)
```

```python
#copiando a sean
pages_parsed=[]
titles=[]
sinopsis=[]

for i in tqdm(range(len(pages))):
    pages_parsed.append(BeautifulSoup(pages[i], 'html.parser'))
    movies_html=pages_parsed[i].select('div.lister-item') #no se si nos ira bien ni de donde slae el div lister
    for m in range(len(movies_html)):
        titles.append(soup.select('h3 > a')[m].get_text(strip=True)) #esta parte me la he sacado yo de la manga
        sinopsis.append(soup.select('p.text-muted')[1+2*m].get_text(strip=True)) # lo mismo
```

## scraping presidents from wikipedia

```python
url='https://en.wikipedia.org/wiki/List_of_presidents_of_the_United_States'
```

```python
response = requests.get(url)
response.status_code
```

```python
soup=BeautifulSoup(response.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
soup.select('td > b > a')
```

```python
len(soup.select('td > b > a'))
```

```python
soup.select('td > b > a')[0].get_text(strip=True)
```

```python
#sean hace
presi_url=soup.select('td > b > a')
```

```python
#para scar url nuevas sean hace

links=[]

for p in presi_url:
    href1=p['href']
    url='https://en.wikipedia.org'+href1
    links.append(url)
    print(url)
```
