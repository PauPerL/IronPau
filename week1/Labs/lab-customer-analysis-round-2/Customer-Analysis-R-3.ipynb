```python
import pandas as pd
#now lets try to build a function that does all of that

def data_deal(file, date_col):
    #we get the file
    df1 = pd.read_csv(file)
    
    #lowercase the columns
    col1 = []

    for i in range(len(df1.columns)):
        col1.append(df1.columns[i].lower())
    
    df1.columns = col1
    
    #we check the info to know which col are numerical and which are categorical
    print("\nInfo of df1\n")
    print(df1.info())
    
    #deal with nan values
    print("\nNulls on col\n")
    print(df1.isnull().sum())
    df1= df1.dropna(subset="customer")
    
    #now lets deal with data columns
    #i was gona do an if to check if date_col is filled but i wont

    df1[date_col] = pd.to_datetime(df1[date_col], errors='coerce')
    
    #make the months col
    #i was also goin to do a new arg to the function to name the new column but i wont as we would have problems with the query
    df1['months']=pd.DatetimeIndex(df1[date_col]).month
    
    #and lets do the motnh query
    dfq1=df1.query('months<=3')
    print("\nData frame after query\n")
    print(dfq1)

#we start the function arg1-str-name of file arg2-str-name of date column
data_deal('marketing_customer_analysis.csv', 'effective to date')
```
